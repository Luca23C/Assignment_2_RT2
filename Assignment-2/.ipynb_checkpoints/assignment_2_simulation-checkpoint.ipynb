```python
import rospy
import time
import sys
import termios
from pynput import keyboard
import actionlib
import assignment_2_2023.msg
from std_msgs.msg import String
from assignment_2_2023.msg import Info
from nav_msgs.msg import Odometry
import ipywidgets as widgets
import matplotlib.pyplot as plt
import tf
from tf.transformations import quaternion_matrix
import numpy as np
from matplotlib.animation import FuncAnimation
from assignment_2_2023.msg import PlanningActionGoal
```

```python
class ActionClient:

    # Define global variable for analize the current status of the target
    target_status = None
    x_target = None
    y_target = None
    # useful for plot information with histogram plot
    target_reached = 0
    target_cancelled = 0


    def __init__(self, histo):
        self.histo = histo
        rospy.init_node('bug_ac')
        self.client = actionlib.SimpleActionClient('reaching_goal', assignment_2_2023.msg.PlanningAction)
        self.client.wait_for_server()
        
        # Create user interface
        self.Title = widgets.HTML(value='<b>Welcome to user interface!</b>')
        self.subtitle = widgets.Label(value='Here you can choose the target position for the robot.')

        # Create text line widget
        self.textline1 = widgets.FloatText(description='<b>x:</b>')
        self.textline2 = widgets.FloatText(description='<b>y:</b>')

        # Create Button widgets
        self.button1 = widgets.Button(description='Send goal')
        self.button2 = widgets.Button(description='Delete goal')
                
        # Create the widgets layout
        self.title_box = widgets.VBox([self.Title, self.subtitle])
        self.float_text_box = widgets.VBox([self.textline1, self.textline2])
        self.button_box = widgets.HBox([self.button1, self.button2])
                
        # Display interface
        self.layout = widgets.VBox([self.title_box, self.float_text_box, self.button_box])
        display(self.layout)
        
        
        # Check if the buttons are pressed
        self.button1.on_click(self.send_goal_button_clicked)
        self.button2.on_click(self.delete_goal_button_clicked)


        # Define subscriber for checking the status of the goal
        self.sub_stat = rospy.Subscriber('/reaching_goal/feedback', String, self.status_callback)

        # Define publisher and subscriber for custom message
        self.sub = rospy.Subscriber('/odom', Odometry, self.odom_callback)
        self.pub = rospy.Publisher('/custom_message', Info, queue_size=10)

        
    def send_goal_button_clicked(self, b):
        self.x_target = self.textline1.value
        self.y_target = self.textline2.value
        self.create_goal(self.x_target, self.y_target)
        
    
    def delete_goal_button_clicked(self, b):
        self.client.cancel_goal()
        time.sleep(2)
        rospy.loginfo(self.target_status)
        
        # Clear textline in the interface
        self.textline1.value = 0
        self.textline2.value = 0
        
        # Update the value of target cancelled
        self.target_cancelled += 1
        
        # Update plot
        self.histo.update_histo(self.target_reached, self.target_cancelled)
    
    
    def status_callback(self, stat):
        self.target_status = stat.feedback.stat
        
        if (self.target_status == 'Target reached!'):
            rospy.loginfo(self.target_status)
            
            # Update the value of target cancelled
            self.target_reached += 1
            
            # Clear textline in the interface
            self.textline1.value = 0
            self.textline2.value = 0
            
            # Update plot
            self.histo.update_histo(self.target_reached, self.target_cancelled)
            
  
    def odom_callback(self, odom_msg):
        # Fill the field of my custom message
        info_msg = Info()
        info_msg.pose_x = odom_msg.pose.pose.position.x
        info_msg.pose_y = odom_msg.pose.pose.position.y
        info_msg.vel_x = odom_msg.twist.twist.linear.x
        info_msg.vel_z = odom_msg.twist.twist.angular.z

        self.pub.publish(info_msg)


    def create_goal(self, x, y):
        # Construction goal message
        goal = assignment_2_2023.msg.PlanningGoal()
        goal.target_pose.pose.position.x = x
        goal.target_pose.pose.position.y = y

        self.client.send_goal(goal)
        rospy.loginfo("Pose goal sent to the action server.")
```

```python
class Visualiser:
    
    y_data = []
    x_data = []
    
    def __init__(self, env):
        self.env = env
        self.sub = rospy.Subscriber('/odom', Odometry, self.odom_callback)
        self.fig, self.ax = plt.subplots()
        self.ln, = plt.plot([], [], 'ro')

        
    def plot_init(self):
        self.ax.set_title('Path navigation rover')
        self.ax.set_xlim(10, -10)
        self.ax.set_ylim(10, -10)
        
        return self.ln,

    
    def odom_callback(self, msg):
        self.y_data.append(msg.pose.pose.position.y)
        self.x_data.append(msg.pose.pose.position.x)
        

    def update_plot(self, frame):
        if (self.env.x_target and self.env.y_target) != None:
            # Allow to visualize the target inside the plot
            self.ax.scatter(self.env.x_target, self.env.y_target, color='blue', s=50)
            self.ln.set_data(self.x_data, self.y_data)
        else:
            self.ln.set_data(self.x_data, self.y_data)
            
        return self.ln,
```

```python
class LiveHistogram:
    
    def __init__(self):
        self.fig, self.ax = plt.subplots()
        self.labels = ['Target reached', 'Target not reached']
        self.x = np.arange(len(self.labels))
        self.bar = self.ax.bar(self.x, [0, 0], color=['green', 'red'])
        
        # Structure of plot histogram
        self.ax.set_ylim(0, 10)
        self.ax.set_xticks(self.x)
        self.ax.set_xticklabels(self.labels)
        self.ax.set_xlabel('Status')
        self.ax.set_ylabel('Numbers')
        self.ax.set_title('Live Histogram of reached/not-reached target')
        
    
    def update_histo(self, target_reached, target_cancelled):
        self.bar[0].set_height(target_reached)
        self.bar[1].set_height(target_cancelled)
```

```python
%matplotlib widget
if __name__ == '__main__':
    try:
        
        histo = LiveHistogram()
        node = ActionClient(histo)
        vis = Visualiser(node)

        # Show animation path
        ani = FuncAnimation(vis.fig, vis.update_plot, init_func=vis.plot_init, cache_frame_data=False)
        
        # Show animation histogram
        #ani_hist = FuncAnimation(histo.fig, histo.update_histo, cache_frame_data=False)
        
        plt.show(block=True)

    except rospy.ROSInterruptException:
        pass
```
